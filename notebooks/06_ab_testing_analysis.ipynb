```python
"""
A/B Testing Analysis

Goal:
- Compare two treatments
- Determine which performs better
"""

import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
```

```python
"""
Load A/B testing dataset.
"""

df = pd.read_csv("../data/interim/marketing_ab_test_standardized.csv")

df.head()
```

```python
"""
Check dataset structure.
"""

print(df.shape)
print(df.columns)
df["test_group"].value_counts()
```

```python
"""
Calculate conversion rate for each group.
"""

conversion_rates = df.groupby("test_group")["converted"].mean()

print(conversion_rates)
```

```python
"""
Check number of users in each group.
"""

df["test_group"].value_counts()
```

```python
"""
Split dataset into two groups.
"""

group_A = df[df["test_group"] == df["test_group"].unique()[0]]["converted"]
group_B = df[df["test_group"] == df["test_group"].unique()[1]]["converted"]
```

```python
"""
Compare average conversion.
"""

print("Group A conversion:", group_A.mean())
print("Group B conversion:", group_B.mean())
```

```python
"""
Perform t-test to check if difference is significant.
"""

t_stat, p_value = ttest_ind(group_A, group_B)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
"""
Final recommendation.
"""

if group_A.mean() > group_B.mean():
    print("Strategy A performs better")
else:
    print("Strategy B performs better")

print("\nRecommendation:")
print("Use the better-performing strategy for high-risk customers identified by the churn model.")
```

```python
"""
Enhanced A/B Testing Summary with Interpretation
"""

conversion_A = group_A.mean()
conversion_B = group_B.mean()
lift = conversion_B - conversion_A

print("=" * 60)
print("A/B TESTING FINAL REPORT")
print("=" * 60)

print("\n--- Performance ---")
print(f"Group A Conversion Rate: {conversion_A:.2%}")
print(f"Group B Conversion Rate: {conversion_B:.2%}")
print(f"Lift (B - A): {lift:.2%}")

print("\n--- Statistical Significance ---")
print(f"P-value: {p_value:.6f}")

if p_value < 0.05:
    print("Conclusion: Statistically significant difference")
else:
    print("Conclusion: Difference is not statistically significant")

print("\n--- Decision ---")
if conversion_A > conversion_B:
    best = "A"
else:
    best = "B"

print(f"Best Strategy: {best}")

print("\n--- Business Insight ---")
print("The analysis shows that one treatment leads to higher user conversion.")
print("Conversion is used as a proxy for engagement, which is linked to retention.")
print("This strategy should be applied to customers identified as high-risk in the churn model.")

print("=" * 60)
```

Strategy A significantly outperforms Strategy B in driving user conversion, with a statistically significant uplift. This indicates that Strategy A is more effective at influencing user engagement, which is a key driver of retention.

## Limitations

The A/B testing dataset is not directly linked to churn outcomes in the banking dataset. Instead, conversion is used as a proxy for engagement, which is assumed to be correlated with retention behavior.

This approach simulates how organizations often combine predictive modeling with experimental insights to inform decision-making in the absence of fully integrated datasets.
