```python
"""
Feature Engineering

Goal:
- Prepare dataset for modeling
- Create useful features
- Remove irrelevant columns
"""

import pandas as pd
```

```python
"""
Load churn dataset.
"""

df = pd.read_csv("../data/interim/bank_churn_standardized.csv")

df.head()
```

```python
"""
Remove columns that do not help prediction.
"""

df = df.drop(columns=["rownumber", "customerid", "surname"])
```

```python
"""
Convert categorical variables into numeric form.
"""

df = pd.get_dummies(df, columns=["geography", "gender"], drop_first=True)
```

```python
"""
Create useful derived features.
"""

# Balance per product
df["balance_per_product"] = df["balance"] / (df["numofproducts"] + 1)

# Credit score per age
df["credit_per_age"] = df["creditscore"] / (df["age"] + 1)

# Tenure per age
df["tenure_per_age"] = df["tenure"] / (df["age"] + 1)
```

```python
"""
Check final dataset.
"""

df.head()
```

```python
"""
Save processed dataset for modeling.
"""

df.to_csv("../data/processed/churn_features.csv", index=False)

print("Saved to processed folder")
```
